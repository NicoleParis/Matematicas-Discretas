# Introducción

Ted es una comedia dirigida por Seth MacFarlane.

La primer película lanzada en 2012, sigue la historia de John Bennett y su osito de peluche viviente Ted. 

La secuela Ted 2, fue estrenada en 2015 y sigue las aventuras de Ted mientras lucha por ser reconocido legalmente como una persona.

<img src=https://www.justwatch.com/images/backdrop/9044576/s640/ted/ted.png>

# Sinopsis 

En Ted, John, un adulto que aún está apegado a su osito de peluche, debe enfrentar los desafíos de su relación con su novia Lori y su deseo de madurar. 

Mientras tanto, Ted lucha por encontrar su lugar en el mundo y evitar caer en manos equivocadas.

Ted 2 continúa explorando la relación entre John y Ted, pero esta vez el enfoque principal está en la lucha legal de Ted para ser reconocido como una persona con derechos legales, lo que conduce a una serie de situaciones cómicas y a menudo absurdas.

# Proyecto 

En este proyecto se realizará un grafo donde se muentren las conexiones entre los personajes de dichas peliculas.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

personajes = [
    "John Bennett", "Ted", "Lori Collins", "Rex", "Donny", "Robert", 
    "Tami-Lynn McCafferty", "Narrador", "Frank", "Guy", "Thomas Skerritt", 
    "Cora", "Jared", "Samantha Leslie Jackson", "Patrick Meighan", "Tom Jessup", 
    "Shep Wild", "Joy", "Axel", "Alex", "Steven Collins"
]

G.add_nodes_from(personajes)

conexiones = [
    ("John Bennett", "Ted"),
    ("John Bennett", "Lori Collins"),
    ("John Bennett", "Rex"),
    ("John Bennett", "Tami-Lynn McCafferty"),
    ("John Bennett", "Narrador"),
    ("John Bennett", "Frank"),
    ("John Bennett", "Guy"),
    ("John Bennett", "Thomas Skerritt"),
    ("John Bennett", "Cora"),
    ("John Bennett", "Jared"),
    ("John Bennett", "Samantha Leslie Jackson"),
    ("John Bennett", "Patrick Meighan"),
    ("John Bennett", "Tom Jessup"),
    ("John Bennett", "Shep Wild"),
    ("John Bennett", "Joy"),
    ("John Bennett", "Axel"),
    ("John Bennett", "Alex"),
    ("John Bennett", "Steven Collins"),
    ("Ted", "Lori Collins"),
    ("Ted", "Rex"),
    ("Ted", "Donny"),
    ("Ted", "Robert"),
    ("Ted", "Tami-Lynn McCafferty"),
    ("Ted", "Narrador"),
    ("Ted", "Frank"),
    ("Ted", "Guy"),
    ("Ted", "Thomas Skerritt"),
    ("Ted", "Cora"),
    ("Ted", "Jared"),
    ("Ted", "Samantha Leslie Jackson"),
    ("Ted", "Patrick Meighan"),
    ("Ted", "Tom Jessup"),
    ("Ted", "Shep Wild"),
    ("Ted", "Joy"),
    ("Ted", "Axel"),
    ("Ted", "Alex"),
    ("Ted", "Steven Collins"),
    ("Lori Collins", "Rex"),
    ("Donny", "Robert"),
    ("Samantha Leslie Jackson", "Patrick Meighan"),
    ("Samantha Leslie Jackson", "Tom Jessup"),
    ("Samantha Leslie Jackson", "Shep Wild"),
    ("Frank", "Axel"),
]

G.add_edges_from(conexiones)

plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)  
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='pink', font_size=10, font_weight='bold', edge_color='black')

plt.title("Conexiones entre personajes de Ted 1 y Ted 2")
plt.show()
```
